```python
import pandas as pd
from yahoofinancials import YahooFinancials
```

```python
import dataframe_image as dfi
```

```python
stocks = ["PG", "WMT", "PEP", "KO", "MDLZ", "COST", "PM", "MO", "CL", "EL", "KMB", "SYY", "STZ", "WBA", "GIS", "K"]
elements = ['trailingEps', 'forwardEps', 'earningsQuarterlyGrowth', 'enterpriseToRevenue', 'enterpriseToEbitda', 'priceToBook', 'pegRatio', 'forwardPE', 'shortPercentOfFloat']
```

```python
data = []
```

```python
print(data)
```

dataframe for many companies and ratios that follows
'trailingEps'
'forwardEps'
'earningsQuarterlyGrowth'
'enterpriseToRevenue'
'enterpriseToEbitda'
'priceToBook'
'pegRatio'
'forwardPE'
'shortPercentOfFloat'

```python
for stock in stocks:    
    stock1 = YahooFinancials(stock)
    statistics = stock1.get_key_statistics_data()
#    print(statistics)
#    print("")
    stat_statement = list(statistics.keys())[0]
#    print(stat_statement)
    statistics = statistics[stat_statement]
#    print(statistics)

    ratios = {"stock": stock,
             "trailingEps" : statistics["trailingEps"],
             'forwardEps': statistics['forwardEps'],
             'earningsQuarterlyGrowth': statistics['earningsQuarterlyGrowth'],
             'enterpriseToRevenue': statistics['enterpriseToRevenue'],
             'enterpriseToEbitda': statistics['enterpriseToEbitda'],
             'priceToBook': statistics['priceToBook'],
             'pegRatio': statistics['pegRatio'],
             'forwardPE': statistics['forwardPE'],
             'shortPercentOfFloat': statistics['shortPercentOfFloat']
             }
    
    print(ratios)
    
    data.append(ratios)
```

```python
df = pd.DataFrame(data, columns = elements, index = stocks)
#df = df.append(data, ignore_index=True)
df
```
