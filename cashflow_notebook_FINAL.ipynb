```python
import pandas as pd
from yahoofinancials import YahooFinancials

%matplotlib inline
```

```python
stocks = 'AAPL'
yf_stocks = YahooFinancials(stocks)
cashflow_info = yf_stocks.get_financial_stmts("quarterly", "cash")
```

```python
dict1 = cashflow_info["cashflowStatementHistoryQuarterly"]["AAPL"]
print(dict1)
```

```python
df0 = pd.DataFrame(dict1[0])
df2 = pd.DataFrame(dict1[1])
df3 = pd.DataFrame(dict1[2])
df4 = pd.DataFrame(dict1[3])
```

```python
print(df0)
```

```python
df0 = df0.transpose()
df0
```

```python
df2 = df2.transpose()
df2
```

```python
df3 = df3.transpose()
df3
```

```python
df4 = df4.transpose()
df4
```

```python
frames = [df4, df3, df2, df0]

result = pd.concat(frames)
data_f = result
```

```python
data_f
```

```python
data_f.reset_index(level=0, inplace=True)
data_f
```

```python
data_f.plot(x="index", y="totalCashFromOperatingActivities")
```
